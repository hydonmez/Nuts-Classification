```python
import os
import numpy as np 
import pandas as pd
import tensorflow as tf
from IPython.display import clear_output as cls 
from keras.preprocessing.image import ImageDataGenerator as IDG 
import plotly.express as px 
import matplotlib.pyplot as plt 
# Model 
from keras.models import Sequential
from tensorflow_hub import KerasLayer as KL
# Optimizer
from tensorflow.keras.optimizers import SGD 
from tensorflow.keras.optimizers.schedules import PiecewiseConstantDecay as PwCD

# Callbacks
from keras.callbacks import EarlyStopping as ES, ModelCheckpoint as MC  
from keras.layers import Conv2D
from keras.models import Sequential
from keras.layers import MaxPooling2D
from keras.layers import Flatten 
from keras.layers import Dense
from keras.layers import Dropout
```

```python

train_path = "C:/Users/hydon/Desktop/Dataset/Train/"
valid_path = "C:/Users/hydon/Desktop/Dataset/Test/"
```

```python
class_names = sorted(os.listdir(train_path))
n_classes = len(class_names)

# Print
print("No. Classes : {}".format(n_classes))
print("Classes     : {}".format(class_names))
```

```python
# Initialize image data generator
train_gen = IDG(rescale=1./255, rotation_range=10, horizontal_flip=True) #image data
valid_gen = IDG(rescale=1./255)

# Load the datasets
train_ds = train_gen.flow_from_directory(train_path, shuffle=True, batch_size=64, target_size=(150, 150), class_mode='sparse')
valid_ds = valid_gen.flow_from_directory(valid_path, shuffle=True, batch_size=32, target_size=(150, 150), class_mode='sparse')
```

```python
model = Sequential()
model.add(InputLayer(input_shape=(150, 150, 3)))

model.add(Conv2D(16, (4,4), activation="relu"))
model.add(Conv2D(16, (4,4), activation="relu")) 
model.add(Conv2D(16, (4,4), activation="relu"))
model.add(Conv2D(16, (4,4), activation="relu")) 
model.add(MaxPooling2D(2,2))


model.add(Conv2D(32, (4,4), activation="relu"))
model.add(Conv2D(32, (4,4), activation="relu")) 
model.add(Conv2D(32, (4,4), activation="relu"))
model.add(Conv2D(32, (4,4), activation="relu")) 
model.add(MaxPooling2D(4,4))

model.add(Conv2D(64, (4,4), activation="relu"))
model.add(Conv2D(64, (4,4), activation="relu")) 
model.add(Conv2D(64, (4,4), activation="relu"))
model.add(Conv2D(64, (4,4), activation="relu")) 

model.add(MaxPooling2D(2,2))


model.add(Flatten())
model.add(Dense(512, activation="relu"))
model.add(Dense(5,activation="softmax"))
```

```python
model_name="Cnn"
model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer="adam",
    metrics=['accuracy']
)
cbs = [ES(patience=7,verbose=1, restore_best_weights=True), MC(model_name+"Cnn"+".h5", save_best_only=True)]
# Training
history = model.fit(train_ds, validation_data=valid_ds, epochs=50, callbacks=cbs)
```
